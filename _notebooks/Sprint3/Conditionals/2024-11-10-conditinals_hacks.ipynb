```python
%%HTML

<output id="combatLog"></output>

<script>
function Character(name, health, attackPower) {
    this.name = name;
    this.health = health;
    this.attackPower = attackPower;
}

// Function for attacking an enemy
function attack(attacker, defender) {
    const damage = Math.floor(Math.random() * attacker.attackPower) + 1; // Random damage up to attackPower
    defender.health -= damage;
    console.log(`${attacker.name} attacks ${defender.name} for ${damage} damage.`);
    document.getElementById("combatLog").innerHTML += `<p>${attacker.name} attacks ${defender.name} for ${damage} damage.</p>`;
}

// Function to check if character is alive
function isAlive(character) {
    return character.health > 0;
}

// Simple turn-based combat function
function battle(player, enemy) {
    console.log(`Battle Start: ${player.name} vs ${enemy.name}`);
    document.getElementById("combatLog").innerHTML += `<p>Battle Start: ${player.name} vs ${enemy.name}</p>`;

    // Combat loop
    while (isAlive(player) && isAlive(enemy)) {
        attack(player, enemy); // Player attacks enemy

        if (!isAlive(enemy)) {
            console.log(`${enemy.name} has been defeated!`);
            document.getElementById("combatLog").innerHTML += `<p>${enemy.name} has been defeated!</p>`;
            break;
        }

        attack(enemy, player); // Enemy attacks player

        if (!isAlive(player)) {
            console.log(`${player.name} has been defeated!`);
            document.getElementById("combatLog").innerHTML += `<p>${player.name} has been defeated!</p>`;
            break;
        }
    }

    console.log("Battle Over");
    document.getElementById("combatLog").innerHTML += `<p>Battle Over</p>`;
}

// Create a player character and an enemy
const player = new Character("Hero", 30, 10);
const enemy = new Character("Baddie", 20, 5);

// Start the battle
battle(player, enemy);
</script>
```
